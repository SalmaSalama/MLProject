```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('data.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
missing_rate = df.isnull().mean().sort_values(ascending=False) * 100
missing_rate = missing_rate[missing_rate > 0]
missing_rate
```

```python
columns_to_drop = missing_rate[missing_rate > 6].index
df_cleaned = df.drop(columns=columns_to_drop)
df_cleaned
```

```python
missing_rate2 = df_cleaned.isnull().mean().sort_values(ascending=False) * 100
missing_rate2 = missing_rate2[missing_rate2 > 0]
missing_rate2
```

```python
df_filled = df_cleaned.copy()
numeric_cols = df_filled.select_dtypes(include=['number']).columns
df_filled[numeric_cols] = df_filled[numeric_cols].fillna(df_filled[numeric_cols].mean())
missing_rate3 = df_filled.isnull().mean().sort_values(ascending=False) * 100
missing_rate3 = missing_rate3[missing_rate3 > 0]
missing_rate3
```

```python
sns.displot(df_filled['SalePrice'],kde='true')
```

```python
sns.boxplot(df_filled['SalePrice'])
```

```python
sns.scatterplot(df_filled['SalePrice'])
```

```python
df_filled['LogSalePrice'] = np.log1p(df_filled['SalePrice'])
plt.figure(figsize=(8, 5))
sns.histplot(df_filled['LogSalePrice'], kde=True, color='mediumseagreen')
```

```python
df_filled['exSalePrice'] = np.sqrt(df_filled['SalePrice'])
plt.figure(figsize=(8, 5))
sns.histplot(df_filled['exSalePrice'], kde=True, color='mediumseagreen')
```
