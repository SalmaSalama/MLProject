# **DBSCAN**

**DBSCAN (Density-Based Spatial Clustering of Applications with Noise) est un algorithme de clustering populaire utilisé en apprentissage automatique et en exploration de données.**

**Il est particulièrement utile pour identifier des clusters de formes et de tailles variées au sein d'un ensemble de données, notamment dans les cas où les clusters peuvent avoir des densités variables ou présenter du bruit (valeurs aberrantes).**

**En principe, toutes les méthodes de clustering utilisent la même approche : nous calculons d'abord les similarités, puis nous les utilisons pour regrouper les points de données en groupes ou lots.**

### **Why do we need a Density-Based clustering algorithm like DBSCAN when we already have K-means clustering?**

- **Gestion du bruit et des valeurs aberrantes** :
- **DBSCAN** est robuste au bruit et aux valeurs aberrantes, car il identifie les points qui ne font partie d'aucun cluster comme des points de bruit.
- **K-Means**, en revanche, attribue chaque point à un cluster, ce qui peut conduire à des résultats de clustering sous-optimaux en présence de valeurs aberrantes.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate synthetic data
X, _ = make_blobs(n_samples=1000, centers=4, random_state=42)

# Visualize Core Data
plt.figure(figsize=(10, 5))

plt.scatter(X[:, 0], X[:, 1], c='black')
plt.title('Blobs Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.show()
```

```python
from sklearn.cluster import KMeans

# K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Visualize K-Means clustering
plt.figure(figsize=(10, 5))

plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', alpha=0.5)
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.show()
```

```python
from sklearn.cluster import DBSCAN

# DBSCAN clustering
dbscan = DBSCAN(eps=0.6, min_samples=6)
dbscan_labels = dbscan.fit_predict(X)

# Visualize DBSCAN clustering
plt.figure(figsize=(12, 5))

plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', alpha=0.5)
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.show()
```

- **Capacité à identifier des formes arbitraires**
- DBSCAN peut identifier des groupes de formes arbitraires, tandis que K-Means suppose que les groupes sont sphériques et isotropes.
- Cela rend DBSCAN plus adapté aux ensembles de données contenant des groupes de formes irrégulières ou de densités variables.

```python
from sklearn.datasets import make_moons

# Generate synthetic data with two moon-shaped clusters
X, _ = make_moons(n_samples=1000, noise=0.06)

# Visualize Core Data
plt.figure(figsize=(10, 5))

plt.scatter(X[:, 0], X[:, 1], c='black')
plt.title('Moons Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.show()
```

```python
# K-Means clustering
kmeans = KMeans(n_clusters=2)
kmeans_labels = kmeans.fit_predict(X)

# Visualize K-Means clustering
plt.figure(figsize=(10, 5))

plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', alpha=0.5)
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.show()
```

```python
# DBSCAN clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Visualize DBSCAN clustering
plt.figure(figsize=(10, 5))

plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', alpha=0.5)
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.show()
```

- **Détermination automatique du nombre de clusters**
- DBSCAN détermine automatiquement le nombre de clusters en fonction des données et des paramètres (« eps » et « min_samples »).
- En revanche, K-Means exige que l'utilisateur spécifie au préalable le nombre de clusters, qui peut ne pas être toujours connu ou varier selon les ensembles de données.

- **Moins sensible à l'initialisation :**
- Le clustering K-Means est sensible au choix initial des centroïdes du cluster, ce qui peut conduire à des résultats finaux différents selon l'initialisation.
- DBSCAN, en revanche, est moins sensible à l'initialisation, car il définit les clusters en fonction de la densité plutôt que des centroïdes.

### **Processus DBSCAN**


1. **Initialisation** :
- Choisissez deux paramètres :
- **ε (epsilon) :** Le rayon du voisinage autour d'un point.
- **MinPts** :** Le nombre minimal de points dans ε requis pour former une région dense.

2. **Identification des points centraux** :
- Pour chaque point de l'ensemble de données :
- Comptez le nombre de points dans son voisinage ε.
- Si le nombre de points dans le voisinage est supérieur ou égal à MinPts, marquez le point comme point central.

3. **Extension des points centraux** :
- Pour chaque point central :
- Créez un nouveau cluster ou développez un cluster existant.
- Trouvez récursivement tous les points directement accessibles depuis le point central dans ε.
- Ajoutez ces points au cluster.
- Si un point est lui-même un point central, répétez le processus pour son voisinage ε.



4. **Affectation des points frontières** :
- Pour chaque point frontière (un point qui n'est pas un point central, mais qui se trouve dans le voisinage ε d'un point central) :
- Affectez le point frontière au cluster de son point central correspondant.

5. **Identification des points de bruit** :
- Tout point qui n'est ni un point central ni un point frontière est considéré comme un point de bruit et n'est affecté à aucun cluster.

6. **Formation des clusters** :
- Répétez le processus d'expansion jusqu'à ce que tous les points aient été visités et affectés à des clusters ou étiquetés comme bruit.

### **DBSCAN Parameter Selection**

**To see changes in parameters, visit this [site](https://www.naftaliharris.com/blog/visualizing-dbscan-clustering/)**

1. **MinPts** :
- MinPts spécifie le nombre minimal de points requis pour former une région dense.
- Il est généralement recommandé que le paramètre minPts ne soit pas inférieur à 3, car DBSCAN exige qu'au moins trois points (y compris le point central) se trouvent dans le voisinage ε de chaque point central.
- Si l'ensemble de données est bruité ou contient des valeurs aberrantes, une valeur minPts plus élevée peut être nécessaire pour garantir un clustering robuste.
2. **eps** :
- Lors du choix du paramètre eps dans DBSCAN, il est essentiel de trouver un équilibre entre l'inclusion d'un nombre suffisant de points dans chaque cluster et l'évitement du sur-clustering ou du sous-clustering.
- Si un eps petit est choisi, de nombreux points peuvent ne pas être attribués à un cluster, ce qui peut entraîner un sous-clustering.
- À l'inverse, si eps est trop grand, de nombreux points peuvent être attribués au même cluster, ce qui peut entraîner un sur-clustering.

- **Vous pouvez identifier le point du coude**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.neighbors import NearestNeighbors

# Function to compute the k-distance plot
def k_distance_plot(X, minPts):
    k = minPts - 1  # k is one less than minPts
    # Compute the distances to the k nearest neighbors for each point
    nbrs = NearestNeighbors(n_neighbors=k).fit(X)
    distances, _ = nbrs.kneighbors(X)
    # Sort the distances
    distances = np.sort(distances[:, -1])  # Select the distances to the k-th neighbor
    return distances

# Generate example data (replace with your own data if needed)
X, _ = make_moons(n_samples=1000, noise=0.05)

# Plot k-distance plot for minPts = 6
distances = k_distance_plot(X, minPts=6)

# We look for the point with the maximum change in distance
diff_distances = np.diff(distances, 2)  # Calculate second order difference
max_change_index = np.argmax(diff_distances) + 1  # Add 1 to account for the offset
eps = distances[max_change_index]  # Get the distance value at the identified index

# Plot the elbow point
plt.plot(distances)
plt.axhline(y=eps, color='red', linestyle='--', label=f'Elbow Point (eps={eps:.2f})')
plt.xlabel('Samples')
plt.ylabel('Distance to 5th Nearest Neighbor')
plt.title('K-Distance Plot (minPts=6)')
plt.legend()
plt.show()

# Perform DBSCAN clustering with chosen eps
dbscan = DBSCAN(eps=eps, min_samples=6)
labels = dbscan.fit_predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('DBSCAN Clustering')
plt.show()
```

### **Avantages :**
- **DBSCAN est capable d'identifier des clusters de forme et de taille arbitraires, ce qui le rend adapté aux jeux de données aux structures complexes.**

- **DBSCAN est robuste au bruit et aux valeurs aberrantes dans les jeux de données. Il y parvient en définissant les clusters comme des régions à forte densité séparées par des régions à faible densité, ignorant ainsi les points de bruit.**

- **DBSCAN détermine automatiquement la densité des clusters en fonction des paramètres *eps* et *minPts*. Il n'est pas nécessaire de spécifier le nombre de clusters au préalable, ce qui peut s'avérer difficile dans de nombreux jeux de données réels.**

### **Inconvénients :**
- **Les performances de DBSCAN dépendent fortement du choix des paramètres, notamment *eps* et *minPts*. La sélection des valeurs optimales pour ces paramètres peut s'avérer complexe et nécessiter des connaissances du domaine ou des expérimentations.**

- **Non déterministe, car les points de bordure accessibles par plusieurs clusters peuvent appartenir à n'importe quel cluster selon l'ordre de traitement des données.**

- **Échec pour les données de très grande dimension en raison du problème de dimensionnalité.**
- **L'algorithme DBSCAN est un peu plus difficile à régler que K-Means.**
